```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
```

```python
dataset = pd.read_csv("buddymove_holidayiq.csv")
df = dataset.copy()
print(df.shape)
df.head(10)
```

```python
X = df.iloc[:,1:].values
sse = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(X)
  sse.append(kmeans.inertia_)


plt.figure()
plt.grid()
plt.plot(range(1,11),sse)
```

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(X)


plt.figure(figsize=(15,7))
sns.scatterplot(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color = 'yellow', label ="Cluster 1")
sns.scatterplot(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color = 'blue', label = "Cluster 2")
sns.scatterplot(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color = 'green', label = "Cluster 3")
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'r', label = 'Centroids',s=300,marker=',')
plt.grid(True)
plt.title('Holiday Dataset')
plt.legend()
plt.show()
```

```python
X = df.iloc[:,1:].values
sse = []
for i in range(1,11):
  kmed = KMedoids(n_clusters = i)
  kmed.fit(X)
  sse.append(kmed.inertia_)


plt.figure()
plt.grid()
plt.plot(range(1,11),sse)
```

```python
kmed = KMedoids(n_clusters = 4)
y_kmed = kmed.fit_predict(X)


plt.figure(figsize=(15,7))
sns.scatterplot(X[y_kmed == 0, 0], X[y_kmed == 0, 1], color = 'yellow', label ="Cluster 1")
sns.scatterplot(X[y_kmed == 1, 0], X[y_kmed == 1, 1], color = 'blue', label = "Cluster 2")
sns.scatterplot(X[y_kmed == 2, 0], X[y_kmed == 2, 1], color = 'green', label = "Cluster 3")
sns.scatterplot(kmed.cluster_centers_[:, 0], kmed.cluster_centers_[:, 1], color = 'r', label = 'Centroids',s=300,marker=',')
plt.grid(True)
plt.title('Holiday Dataset')
plt.legend()
plt.show()
```
