# Polynomial Linear Regression

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

### Training the Linear Regression Model for the whole dataset

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

### Training the Polynomial Regression Model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
pol_reg =PolynomialFeatures(degree =5)
X_poly = pol_reg.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly,y)
```

### Visualising the results of Linear Regression

```python
plt.title("Simple Linear Regression Results")
plt.xlabel("Position of the Employee")
plt.ylabel("Salary of the Employee (in Lakhs)")
plt.scatter(X,y,color='red',marker='x')
plt.plot(X,lin_reg.predict(X))
plt.show()
```

### Visualising the results of Polynomial Regression

```python
plt.title("Polynomial Linear Regression Results")
plt.xlabel("Position of the Employee")
plt.ylabel("Salary of the Employee (in Lakhs)")
plt.scatter(X,y,color='red',marker='x')
plt.plot(X,poly_reg.predict(pol_reg.fit_transform(X)))
plt.show()
```

```python
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape((len(X_grid)),1)
plt.title("Polynomial Linear Regression Results")
plt.xlabel("Position of the Employee")
plt.ylabel("Salary of the Employee (in Lakhs)")
plt.scatter(X,y,color='red',marker='x')
plt.plot(X_grid,poly_reg.predict(pol_reg.fit_transform(X_grid)))
plt.show()
```
